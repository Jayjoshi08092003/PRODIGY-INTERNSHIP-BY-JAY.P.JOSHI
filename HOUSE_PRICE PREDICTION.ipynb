```python
import pandas as pd

# Load the first dataset
df1 = pd.read_csv('train.csv')

# Specify the columns you want to keep from the first dataset
columns_to_keep = ['LotArea', 'BedroomAbvGr', 'FullBath', 'HalfBath']

# Create a new DataFrame with only the specified columns
new_df = df1[columns_to_keep]

# Create a new column 'TotalBath' by summing 'FullBath' and 'HalfBath'
new_df['TotalBath'] = new_df['FullBath'] + new_df['HalfBath']

# Drop the original 'FullBath' and 'HalfBath' columns
new_df = new_df.drop(columns=['FullBath', 'HalfBath'])

# Load the second dataset
df2 = pd.read_csv('sample_submission.csv')

# Extract the 'price' column from the second dataset
price_column = df2[['SalePrice']]

# Optionally, you may need to ensure that the indices of both DataFrames match
# Assuming they have the same number of rows and should be aligned by their indices

# Add the 'price' column to the new DataFrame
new_df['SalePrice'] = price_column

# Display the new DataFrame
print(new_df)
```

```python
new_df.isnull().sum()
```

```python
df2 = new_df.dropna()
df2
```

```python
df2.isnull().sum()
df2.duplicated().sum()
df2.describe()
```

```python

# Define your features and target variable
X = df2[['LotArea', 'BedroomAbvGr', 'TotalBath']]
y = df2['SalePrice']

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=100)
```

```python
my_model = LinearRegression()
```

```python
my_model.fit(X_train, y_train)
```

```python
y_pred = my_model.predict(X_test)
```

```python
print('Coefficients:', my_model.coef_)
print('Intercept:', my_model.intercept_)
```
